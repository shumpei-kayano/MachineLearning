## 質的データをすべてダミー化してデータ分析させるデータフレームを作成。次に、作成したデータフレームの先頭５行を表示。

```python
dmy = pd.get_dummies(df[col],drop_first=True)
df2 = pd.concat([df, dmy],axis = 1)
df2 = df2.drop(col, axis=1)
df2.head(5)
```

## 特徴量xと正解データtに分類
訓練データとテストデータの比率は８：２、乱数シードは０

```python
#特徴量として利用する列のリスト
#col = ['cap-shape','cap-surface','cap-color','odor','gill-attachment','gill-spacing','gill-size','gill-color','stalk-shape','stalk-root']
col2 = ['cap-shape_conical','cap-shape_convex','cap-shape_flat','cap-shape_knobbed','cap-shape_sunken','cap-surface_groovesmooth',
        'cap-surface_scaly','cap-surface_smooth','gill-attachment_free','cap-color_buff','cap-color_cinnamon','cap-color_gray','cap-color_green',
        'cap-color_pink','cap-color_purple','cap-color_red','cap-color_white','cap-color_yellow','gill-spacing_crowded','gill-size_narrow','stalk-shape_tapering',
        'gill-color_chocolate','gill-color_gray','gill-color_green','gill-color_orange','gill-color_pink','gill-color_purple','gill-color_red','gill-color_white',
        'gill-color_yellow','stalk-root_club','stalk-root_equal','stalk-root_missing','stalk-root_rooted'
        ]

x = df2[col2]
t = df2['class_poisonous']

#訓練データとテストデータに分類
x_train,x_test,y_train,y_test = train_test_split(x,t,test_size = 0.2, random_state = 0)
x_train.shape
```

## モデルの作成と学習を実施

```python
#訓練データでの再学習
#model = tree.DecisionTreeClassifier(max_depth = 2, random_state = 0)
#model.fit(x,t)
#model.score(x,t)

#テストデータの予測結果と実際の答えが合致する正解率を計算

model = tree.DecisionTreeClassifier(max_depth = 5,random_state = 0, class_weight = 'balanced')
model.fit(x_train,y_train)

```

##正解率の計算

```python
#正解率を計算
model.score(X = x_test,y = y_test)
```
